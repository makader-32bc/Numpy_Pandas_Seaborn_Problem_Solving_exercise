### Exercise 2: Statistical Operations
### Task: Modify the code to compute the median and standard deviation instead of the mean, and 
### apply it to a 1D array of 20 random floats between 0 and 10. 

```python
import numpy as np
#original Code
array=np.random.randint(1,100,size=10)
mean_value=np.mean(array)
print("Array:",array)
print("Mean:",mean_value)
```

```python
# modification 
array=np.random.uniform(0,10,size=20) #uniform() generate randam float number
median_value=np.median(array)
standard_deviation_value=np.std(array)
print("Array:",array)
print("Median:",round(median_value,2))
print("Standard Deviation:",round(standard_deviation_value,2))
```

### Exercise 3: DataFrame Filtering 
### Task: Modify the code to filter rows where age is greater than 25 and city is "New York".

```python
import pandas as pd

#original code

data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'age': [25, 30, 35, 20],
    'city': ['New York', 'Los Angeles', 'New York', 'Chicago']
      }
df = pd.DataFrame(data)
filtered_df = df[df['age'] > 30]
print("Filtered DataFrame:\n", filtered_df)
```

```python
# Update the filtering condition to select rows where age > 25 and city == "New York". 
df = pd.DataFrame(data)
filtered_df = df[(df['age'] > 25) & (df['city'] == 'New York')]
print("Filtered DataFrame:\n", filtered_df)
```

### Exercise 4: Grouping and Aggregation 
### Task: Modify the code to group by city and compute the maximum age for each city. 

```python

# Original code
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'age': [25, 30, 35, 20, 28],
    'city': ['New York', 'Los Angeles', 'New York', 'Chicago', 'Los Angeles']
}

df = pd.DataFrame(data)

# Group by 'city' and calculate the mean of 'age'
grouped = df.groupby('city')['age'].mean()

print("Grouped means:\n", grouped)
```

```python
# Change the aggregation function from mean() to max(). 
# Print the maximum age for each city.

df = pd.DataFrame(data)

# Group by 'city' and calculate the max of 'age'
grouped = df.groupby('city')['age'].max()

print("Grouped max age\n", grouped)
```

### Exercise 6: Correlation Heatmap 
### Task: Modify the code to create a heatmap of the correlation matrix for numerical columns in the tips dataset, with annotations.

```python
# Original code
tips=sns.load_dataset('tips')
sns.heatmap(tips.corr(numeric_only=True))
plt.show()
```

```python

"""
Instructions: 
Ensure the correlation matrix includes only numerical columns (use select_dtypes or 
specify columns). 
Add annotations to the heatmap to show correlation values (use annot=True). 
Set the color map to coolwarm for better visualization.
"""
# modified code
tips=sns.load_dataset('tips')

#Select only numeric columns
numeric_tips = tips.select_dtypes(include='number')
corr_matrix=numeric_tips.corr()
# Add annotations to the heatmap
sns.heatmap(corr_matrix, annot=True)
plt.show()
# add colour of the map
sns.heatmap(corr_matrix, annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap of Numeric Columns in Tips Dataset")
plt.show()
```

# Plotly Exercises 

### Exercise 7: Interactive Scatter Plot 
### Task: Modify the code to create a scatter plot of total_bill vs. tip, with points colored by day and sized by size. 


```python
import plotly.express as px
#original code
tips=px.data.tips()
fig=px.scatter(tips, x='total_bill',y='tip',color='sex')
fig.show()
```

```python
"""
Instructions: 
 Change the color parameter to day. 
 Add the size parameter to scale points by the size column. 
 Add hover data to show time and smoker when hovering over points. 
"""
#modified code
fig = px.scatter(
    tips,
    x='total_bill',        # x-axis
    y='tip',               # y-axis
    color='day',           # Color points by day of the week
    size='size',           # Scale points by party size
    hover_data=['time', 'smoker'],  # Show additional info on hover
    title="Scatter Plot of Total Bill vs Tip by Day"
)

# Display the plot
fig.show()
```

### Exercise 8: Bar Plot with Customization 
### Task: Modify the code to create a bar plot showing the average total_bill per day, with bars colored by sex. 

```python
import plotly.express as px
#original code
tips=px.data.tips()
fig=px.bar(tips, y='total_bill',x='day')
fig.show()
```

```python
"""
Instructions: 
 Modify the code to aggregate total_bill by day to show the mean (use 
tips.groupby('day')['total_bill'].mean() or Plotly’s built-in aggregation). 
 Add the color parameter to differentiate bars by sex. 
 Update the layout to include a title: "Average Total Bill by Day and Sex".

"""

fig = px.bar(
    tips,
    x='day',                # x-axis: day of the week
    y='total_bill',          # y-axis: total bill
    color='sex',             # separate bars by sex
    barmode='group',         # group bars side by side
    title="Average Total Bill by Day and Sex", 
    labels={'total_bill': 'Average Total Bill'}  # optional y-axis label
)

# Display the plot
fig.show()
```
